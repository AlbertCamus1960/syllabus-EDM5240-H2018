```python
huff = pan.read_csv("huffpost.csv", low_memory=False)
```

```python
huff
```

```python
huff.byline.value_counts().head(15)
```

```python
signeHP = huff.byline.str.contains("[H,h][U,u][F,f][F,f,i,P,p]|[A,a][O,o][L,l]")
huff[signeHP].shape
```

```python
nonSigneHP = ~huff.byline.str.contains("[H,h][U,u][F,f][F,f,i,P,p]|[A,a][O,o][L,l]")
huff[nonSigneHP].shape
```

```python
huff.pays.value_counts()
```

```python
huff.pays.value_counts().plot()
```

```python
huff.pays.value_counts().plot(kind="pie",title="Éditions inter du HP")
```

```python
huff.pays.value_counts().plot(kind="barh",title="Éditions inter du HP",color="green")
```
